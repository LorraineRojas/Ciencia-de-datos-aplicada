#### Top 5 atributos importantes para el análisis inmobiliario:

### 2. Análisis univariado

####Precio

```python
# Limpiar la columna de precios
data["price_float"] = data["price"].str.replace("[$,]", "", regex=True).astype(float)
```

```python
precio_stats = {
    'Promedio': data["price_float"].mean(),
    'Mínimo': data["price_float"].min(),
    'Máximo': data["price_float"].max(),
    'Mediana': data["price_float"].median(),
    'Q1 (25%)': data["price_float"].quantile(0.25),
    'Q3 (75%)': data["price_float"].quantile(0.75)
}

print("Estadísticas del Precio (price):")
for k, v in precio_stats.items():
    print(f"{k}: ${v:,.2f}")
```

```python
# Histograma de Precios
plt.figure(figsize=(12, 6))
sns.histplot(data["price_float"], bins=30, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precio ($)')
plt.ylabel('Frecuencia')
plt.show()
```

####2. Barrio

```python
# Valores únicos de los vecindarios.
data["neighbourhood_cleansed"].unique()
```

```python
# Frecuencia de Barrios (neighbourhood)
neighbourhood_freq = data['neighbourhood_cleansed'].value_counts().head(5)

# Gráfico de Barras para Barrios más frecuentes
plt.figure(figsize=(10, 5))
sns.barplot(x=neighbourhood_freq.index, y=neighbourhood_freq.values)
plt.title('Barrios más Frecuentes')
plt.xlabel('Barrio')
plt.ylabel('Número de Propiedades')
plt.xticks(rotation=45)
plt.show()
```

```python
copacabana_count = data[data['neighbourhood_cleansed'] == 'Copacabana'].shape[0]
print(f"Propiedades en Copacabana: {copacabana_count}")
```

#### 3. Tipos de habitación

```python
# Frecuencia de Tipos de habitación (room_type)
data["room_type"].value_counts(dropna=False, normalize=True)
```

#### 4. Disponibilidad anual

```python
availability_stats = {
    'Promedio': data['availability_365'].mean(),
    'Mediana': data['availability_365'].median(),
    'Mínimo': data['availability_365'].min(),
    'Máximo': data['availability_365'].max()
}

print("\nEstadísticas de Disponibilidad (availability_365):")
for k, v in availability_stats.items():
    print(f"{k}: {v:.2f} días")
```

####5. Calificaciones o reseñas

```python
# Estadísticas de Calificación en reseñas (review_scores_rating)
rating_stats = {
    'Promedio': data['review_scores_rating'].mean(),
    'Mediana': data['review_scores_rating'].median(),
    'Mínimo': data['review_scores_rating'].min(),
    'Máximo': data['review_scores_rating'].max()
}

print("\nEstadísticas de Calificación en Reseñas (review_scores_rating):")
for k, v in rating_stats.items():
    print(f"{k}: {v:.2f} / 5")
```

```python
# Histograma de Calificaciones
plt.figure(figsize=(12, 6))
sns.histplot(data['review_scores_rating'], bins=20, kde=True)
plt.title('Distribución de Calificaciones de Reseñas')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show()
```

### 3.  Estrategia de análisis

La estrategia de análisis se basará en un enfoque mixto entre el análisis exploratorio de datos (EDA) y técnicas estadísticas descriptivas. El objetivo es identificar las mejores opciones de inversión inmobiliaria en función de variables clave como el precio, la ubicación, el tipo de propiedad y la disponibilidad. Para ello:

* Cálculo de estadísticos básicos: Se calcularán medidas de tendencia central (media, mediana) y de dispersión (desviación estándar, percentiles) para el precio, la disponibilidad, y las calificaciones de reseñas.
* Técnicas gráficas: Se usarán gráficos de dispersión (scatterplots) para observar la relación entre precio y otros factores como la ubicación y el tipo de propiedad. Además, se utilizarán histogramas para analizar la distribución de precios y reseñas, y un mapa de calor para visualizar correlaciones entre variables clave.
* Justificación: El análisis exploratorio de datos nos permitirá entender la distribución y relación entre variables clave, identificando así los mejores sectores y propiedades para invertir basados en datos objetivos. La visualización facilitará la identificación de patrones y tendencias que no son evidentes solo con estadísticas.

###4. Desarrollo de la estrategia

```python
# Procesar datos faltantes
data['review_scores_rating'] = data['review_scores_rating'].fillna(data['review_scores_rating'].mean())
```

```python
# Gráfico de dispersión (precio vs calificación)
plt.figure(figsize=(10,6))
sns.scatterplot(x="price_float", y='review_scores_rating', data=data)
plt.title('Relación Precio vs Calificación')
plt.show()
```

```python
# Agrupar datos por barrio
barrio_group = data.groupby('neighbourhood')["price_float"].mean().sort_values(ascending=False).head(10)
print("Barrios con precios más altos:")
print(barrio_group)
```

```python
# Mapa de calor de correlaciones
plt.figure(figsize=(12,8))
sns.heatmap(data[["price_float", 'availability_365', 'review_scores_rating']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```

###5. Generación de resultados

####Recomendaciones para Inversión Inmobiliaria

En este análisis de propiedades listadas en la ciudad de Río de Janeiro, se evaluaron varias características clave para identificar las mejores oportunidades de inversión inmobiliaria. El enfoque estuvo centrado en barrios estratégicos, tipos de propiedad, precios y disponibilidad, con el objetivo de maximizar el retorno de inversión basado en datos.

1. Barrios Recomendados:
Los barrios de Copacabana e Ipanema destacan como las mejores opciones para inversión inmobiliaria. Copacabana cuenta con un volumen significativo de propiedades (1,867 listadas), lo que refleja una alta demanda en la zona. Ipanema, aunque con un número menor de propiedades, tiene precios y calificaciones favorables, lo que la convierte en una opción premium para inversiones de alto retorno.

2. Tipo de Propiedad:
Se recomienda enfocar la inversión en propiedades de tipo "Entire home/apt". Estas propiedades representan el 79% de las ofertas disponibles y tienen una mayor tasa de ocupación y satisfacción del cliente, medida a través de las calificaciones de reseñas. Este tipo de propiedad es preferido por los turistas, ya que ofrece mayor privacidad y flexibilidad, lo que incrementa su valor en el mercado.

3. Rango de Precios:
El análisis de precios muestra que las propiedades más rentables se encuentran en un rango de entre 200 y 600 dolares por noche. Este intervalo representa el punto óptimo entre accesibilidad y rentabilidad. Propiedades fuera de este rango, especialmente aquellas con precios excesivamente altos o bajos, tienden a tener una menor demanda o presentan inconsistencias en la calidad, según las reseñas.

4. Disponibilidad y Calificaciones:
Las propiedades con alta disponibilidad (promedio de 203 días por año) y calificaciones de reseñas superiores a 4.75 sobre 5 son las más recomendables para inversión. Esto indica un alto nivel de satisfacción del cliente y la capacidad de generar ingresos constantes a lo largo del año. Invertir en propiedades con alta disponibilidad y buena reputación proporciona una mayor estabilidad financiera y asegura un retorno constante.

####Conclusión
A partir de los datos analizados, se recomienda a los inversionistas concentrarse en los barrios de Copacabana e Ipanema, en propiedades completas con precios dentro del rango de 200 a 600 dólares por noche. Estas áreas y tipos de propiedad ofrecen el equilibrio ideal entre demanda, rentabilidad y satisfacción del cliente, asegurando una inversión sólida y rentable a largo plazo.
